```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data= pd.read_csv('vgsales.csv')
data
```

 Bar Chart 

```python
data_count=data['Platform'].value_counts().head()
names=data_count.index
count=data_count.values
names
```

```python

plt.bar(names, count ,color=['#08306b','#084b8a','#0868ac','#9ecae1','#deebf7'])
plt.xlabel('platform')
plt.ylabel('Number of games release')
plt.title('Platform vs Number of games released')
plt.show()
```

pie chart 

```python
games_genre=data['Genre'].value_counts().head()
type=games_genre.index
count=games_genre.values

plt.pie(games_genre ,labels=type,autopct='%1.3f%%',colors=['#3f007d','#54278f','#807dba','#9e9ac8','#dadaeb'] )
plt.title('Video games genre distripution')
plt.show()
```

Histgrame : years of release 

```python
mydata=data[data['Year'] > 2015]
plt.hist(mydata['Year'] , bins= 20 , color='skyblue' , edgecolor='white')
```

```python
data['Global_Sales']
```

```python
plt.hist(data['Global_Sales'], bins=50, linewidth=0.5, color='skyblue', edgecolor='white', log=True)
plt.xlim(0,60)
plt.xlabel('Global sale')
plt.ylabel('games sales count')
plt.show()
```

boxplot : games genre and it is global sale 

```python
plt.figure(figsize=(15, 6))  # Increase width to spread categories

sns.boxplot(x=data['Genre'] , y=data['Global_Sales'],palette="coolwarm")
plt.xticks(rotation=45)
plt.yscale("log")
plt.xlabel('Game genre')
plt.ylabel('Global Sale')
plt.title('Global sale distripution for each game genre')
plt.show()
```

scatter 

```python
plt.figure(figsize=(15, 6))
plt.scatter(data['Genre'] , data['Global_Sales'] , c=data['Year'], alpha=0.5, cmap="viridis")
plt.xticks(rotation=45)
# plt.yscale('log')
plt.colorbar(label="year of ralease")

plt.show()
```

Line 

```python
avg_sale_per_year = data.groupby('Year')['Global_Sales'].mean()
plt.plot(avg_sale_per_year.index , avg_sale_per_year , linestyle='-' )
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Year vs Global Sale')
plt.show()
```
